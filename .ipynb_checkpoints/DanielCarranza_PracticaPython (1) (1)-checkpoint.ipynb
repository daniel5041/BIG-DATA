## Ejercicio 3: Calculadora básica
**Enunciado:** Definir dos números y una operación matemática. Las operaciones permitidas son suma, resta, multiplicación y división. Se debe validar que no se divida entre cero.

```python
# Definir los dos números y la operación a realizar
num1 = 10
num2 = 5
operacion = "division"

# Según la operación elegida se hace el cálculo correspondiente
if operacion == "suma":
    resultado = num1 + num2
elif operacion == "resta":
    resultado = num1 - num2
elif operacion == "multiplicacion":
    resultado = num1 * num2
elif operacion == "division":
    if num2 != 0:
        resultado = num1 / num2
    else:
        resultado = "Error: no se puede dividir entre 0"
else:
    resultado = "Operación no válida"

# Mostramos los resultados
print("Número 1:", num1)
print("Número 2:", num2)
print("Operación:", operacion)
print("Resultado:", resultado)
```

**Explicación:**
Se guardaron los dos números y la operación en variables. Con if y elif se revisa cuál operación fue elegida y se hace el cálculo. En el caso de la división se verifica primero que el segundo número no sea cero para evitar un error, y si lo es se muestra un mensaje en lugar del resultado.

## Ejercicio 4: Número par o impar
**Enunciado:** Definir un número entero, determinar si es par o impar y mostrar un mensaje con el resultado.

```python
# Definir el numero 
numero = 17

# Usamos el % para obtener el residuo de dividir entre 2 el numero de la variable
# Si el resultado es 0, el número es par y si es 1, es impar
if numero % 2 == 0:
    print(f"El número {numero} es PAR.")
else:
    print(f"El número {numero} es IMPAR.")
```

**Explicación:**
El operador `%` devuelve el residuo de una división. Al dividir cualquier número entre 2, si el resultado es `0` significa que el número es divisible entre 2, por lo tanto es **par**. Si el residuo es `1`, no es divisible exactamente entre 2, por lo tanto es **impar**. Se utilizó una estructura condicional `if / else` para mostrar el mensaje correspondiente según el resultado.

## Ejercicio 5: Tabla de multiplicar
**Enunciado:** Definir un número entero y mostrar su tabla de multiplicar del 1 al 12 usando un ciclo.

```python
# Definir el numero del que queremos la tabla
numero = 7

# Mostramos un título para saber de qué tabla se trata
print("Tabla de multiplicar del", numero)

# Este ciclo se repite 12 veces, una por cada número del 1 al 12
for i in range(1, 13):
    
    # Multiplicamos el número por i que va cambiando
    resultado = numero * i
    
    # Mostramos la operación y su resultado
    print(numero, "x", i, "=", resultado)
```

**Explicación:**
Para hacer la tabla se utilizó un ciclo for que repite el código 12 veces, una por cada número del 1 al 12. Se usa range del 1 al 13 porque así Python empieza en 1 y llega hasta 12, el 13 no aparece porque Python siempre excluye el último número que se le da. Dentro del ciclo, en cada repetición la variable i va cambiando de 1, 2, 3... hasta 12 y se multiplica por el número de la tabla. Al final se muestra en el resultado con un print.

## Ejercicio 6: Contador de vocales
**Enunciado:** Definir una frase y contar cuántas vocales contiene. El programa debe funcionar con mayúsculas y minúsculas.

```python
# Definir la frase que vamos a analizar
frase = "Programar En Python Es Muy Interesante"

# Escribimos todas las vocales en minusculas y mayusculas 
vocales = "aeiouAEIOU"

# Creamos un contador que empieza en cero
contador = 0

# Recorremos la frase letra por letra
for letra in frase:
    # Si la letra actual es una vocal, sumamos 1 al contador
    if letra in vocales:
        contador = contador + 1

# Mostramos los resultados
print("Frase:", frase)
print("Total de vocales:", contador)
```

## Ejercicio 8: Lista de números positivos y negativos
**Enunciado:** Definir 10 números, contar cuántos son positivos, negativos y ceros, y mostrar un resumen final.

```python
# Definir los 10 números en una lista
numeros = [4, -3, 0, 7, -1, 0, 9, -5, 2, -8]

# Contadores para cada caso
positivos = 0
negativos = 0
ceros = 0

# Recorremos cada número de la lista
for numero in numeros:
    if numero > 0:
        positivos = positivos + 1
    elif numero < 0:
        negativos = negativos + 1
    else:
        ceros = ceros + 1

# Mostramos el resumen
print("Números analizados:", numeros)
print("Positivos:", positivos)
print("Negativos:", negativos)
print("Ceros:", ceros)
```

**Explicación:**
Se guardaron los 10 números en una lista y se crearon 3 contadores en cero. Con un ciclo `for` se revisó cada número y según su valor se sumó 1 al contador que correspondía. Al final se imprimió el resumen de cada grupo.

## Ejercicio 11: Función para calcular descuento
**Enunciado:** Crear una función llamada `calcular_descuento` que reciba el precio original y el porcentaje de descuento y retorne el precio final.

```python
# Crear la función que calcula el descuento
def calcular_descuento(precio_original, porcentaje):
    # Calculamos cuanto es el descuento
    descuento = precio_original * porcentaje / 100
    # Restamos el descuento al precio original
    precio_final = precio_original - descuento
    return precio_final

# Definimos los valores
precio = 500
descuento = 20

# Llamamos la función y guardamos el resultado
resultado = calcular_descuento(precio, descuento)

# Mostramos los resultados
print("Precio original:", precio)
print("Descuento:", descuento, "%")
print("Precio final:", resultado)
```

**Explicación:**
La función recibe dos datos: el precio original y el porcentaje de descuento. Adentro calcula cuánto es ese porcentaje del precio y lo resta. Con `return` devuelve el precio final para poder usarlo fuera de la función.

## Ejercicio 12: Validación de contraseña
**Enunciado:** Validar una contraseña definida en el código. Debe tener al menos 8 caracteres, un número, una mayúscula y una minúscula. Indicar si es válida o no.

```python
# Definir la contraseña a validar
contrasena = "Python2024"

# Variables para verificar cada condición
tiene_8_caracteres = len(contrasena) >= 8
tiene_numero = False
tiene_mayuscula = False
tiene_minuscula = False

# Revisamos cada carácter de la contraseña
for caracter in contrasena:
    if caracter.isdigit():
        tiene_numero = True
    if caracter.isupper():
        tiene_mayuscula = True
    if caracter.islower():
        tiene_minuscula = True

# Verificamos si cumple todas las condiciones
if tiene_8_caracteres and tiene_numero and tiene_mayuscula and tiene_minuscula:
    print("La contraseña", contrasena, "es VÁLIDA")
else:
    print("La contraseña", contrasena, "NO es válida")
    if not tiene_8_caracteres:
        print("- Debe tener al menos 8 caracteres")
    if not tiene_numero:
        print("- Debe contener al menos un número")
    if not tiene_mayuscula:
        print("- Debe contener al menos una mayúscula")
    if not tiene_minuscula:
        print("- Debe contener al menos una minúscula")
```

**Explicación:**
Se usó `len()` para contar los caracteres. Con un ciclo `for` se revisó cada letra usando `isdigit()` para detectar números, `isupper()` para mayúsculas y `islower()` para minúsculas. Si todas las condiciones se cumplen la contraseña es válida, si no se indica cuál condición falla.

## Ejercicio 13: Sistema simple de inventario
**Enunciado:** Crear una lista de productos con código, nombre, precio y cantidad. Mostrar todos los productos, buscar uno por código y calcular el valor total del inventario.

```python
# Crear la lista de productos como diccionarios
productos = [
    {"codigo": 1, "nombre": "Cuaderno", "precio": 25, "cantidad": 10},
    {"codigo": 2, "nombre": "Lapicero", "precio": 5, "cantidad": 50},
    {"codigo": 3, "nombre": "Mochila", "precio": 150, "cantidad": 8},
    {"codigo": 4, "nombre": "Regla", "precio": 10, "cantidad": 30}
]

# Mostrar todos los productos
print("Lista de productos:")
for producto in productos:
    print("Código:", producto["codigo"], "| Nombre:", producto["nombre"], "| Precio:", producto["precio"], "| Cantidad:", producto["cantidad"])

# Buscamos un producto por código
codigo_buscar = 3
print("\nBuscando producto con código", codigo_buscar)
for producto in productos:
    if producto["codigo"] == codigo_buscar:
        print("Producto encontrado:", producto["nombre"])

# Calculamos el valor total del inventario
total = 0
for producto in productos:
    total = total + producto["precio"] * producto["cantidad"]
print("\nValor total del inventario:", total)
```

**Explicación:**
Cada producto se guardó como un diccionario dentro de una lista. Se usaron ciclos `for` para recorrer la lista y mostrar los productos, buscar por código comparando con `==`, y calcular el total multiplicando precio por cantidad en cada producto.

## Ejercicio 14: Registro de ventas
**Enunciado:** Crear una lista de ventas con cliente, producto, cantidad y precio unitario. Calcular el total por venta, el total general e identificar el cliente con la compra más alta.

```python
# Creamos la lista de ventas
ventas = [
    {"cliente": "Ana García", "producto": "Mochila", "cantidad": 2, "precio_unitario": 150},
    {"cliente": "Carlos Pérez", "producto": "Cuaderno", "cantidad": 5, "precio_unitario": 25},
    {"cliente": "María López", "producto": "Lapicero", "cantidad": 10, "precio_unitario": 5},
    {"cliente": "José Martínez", "producto": "Mochila", "cantidad": 3, "precio_unitario": 150}
]

# Variables para el total general y la venta más alta
total_general = 0
venta_mas_alta = 0
cliente_mayor_compra = ""

print("Detalle de ventas:")
for venta in ventas:
    # Calculamos el total de cada venta
    total_venta = venta["cantidad"] * venta["precio_unitario"]
    total_general = total_general + total_venta
    print("Cliente:", venta["cliente"], "| Producto:", venta["producto"], "| Total:", total_venta)

    # Verificamos si esta venta es la más alta
    if total_venta > venta_mas_alta:
        venta_mas_alta = total_venta
        cliente_mayor_compra = venta["cliente"]

print("\nTotal general vendido:", total_general)
print("Cliente con la compra más alta:", cliente_mayor_compra, "con", venta_mas_alta)
```

**Explicación:**
Cada venta se guardó como diccionario en una lista. Con un ciclo `for` se calculó el total de cada venta multiplicando cantidad por precio. Se fue comparando cada total contra el mayor encontrado hasta ese momento para identificar al cliente con la compra más alta.

## Ejercicio 15: Limpieza básica de texto
**Enunciado:** Definir una frase, eliminar espacios al inicio y al final, convertirla a minúsculas, reemplazar espacios dobles por uno solo, contar las palabras y explicar los métodos `strip()`, `lower()`, `replace()` y `split()`.

```python
# Definimos la frase con espacios extra
frase = "  La programacion   en Python   es interesante  "

# Eliminamos espacios al inicio y al final
frase = frase.strip()

# Convertimos todo a minusculas
frase = frase.lower()

# Reemplazamos espacios dobles por uno solo
frase = frase.replace("   ", " ")
frase = frase.replace("  ", " ")

# Contamos las palabras
palabras = frase.split()
cantidad_palabras = len(palabras)

# Mostramos los resultados
print("Frase limpia:", frase)
print("Cantidad de palabras:", cantidad_palabras)
```

**Explicación:**
`strip()` elimina los espacios que sobran al inicio y al final. `lower()` convierte todas las letras a minúsculas. `replace()` busca un texto dentro de la frase y lo cambia por otro, en este caso espacios dobles por uno solo. `split()` divide la frase en una lista de palabras separándolas por los espacios, y con `len()` se cuenta cuántas palabras quedaron.

```python
## Ejercicio 16: DataFrame de estudiantes
**Enunciado:** Crear un DataFrame con al menos 10 estudiantes que incluya nombre, carrera, edad, promedio y ciudad. Mostrar 
las primeras 5 filas, la información general, el promedio general, los estudiantes con promedio mayor o igual a 80 y ordenarlos 
de mayor a menor promedio.
```

```python


# Importamos pandas para poder usar DataFrames
import pandas as pd

# Creamos los datos de los estudiantes en un diccionario
datos = {
    "nombre": [
        "Ana López", "Carlos Pérez", "María García", "José Martínez", "Laura Díaz",
        "Pedro Ramírez", "Sofía Torres", "Luis Hernández", "Valeria Cruz", "Diego Flores"
    ],
    "carrera": [
        "Sistemas", "Contabilidad", "Derecho", "Medicina", "Sistemas",
        "Derecho", "Medicina", "Contabilidad", "Sistemas", "Medicina"
    ],
    "edad": [20, 22, 21, 23, 20, 24, 21, 22, 20, 23],
    "promedio": [85, 72, 90, 65, 88, 78, 95, 70, 83, 60],
    "ciudad": [
        "Tegucigalpa", "San Pedro Sula", "Tegucigalpa", "La Ceiba", "Comayagua",
        "San Pedro Sula", "Tegucigalpa", "La Ceiba", "Comayagua", "Tegucigalpa"
    ]
}

# Convertimos el diccionario en un DataFrame
df = pd.DataFrame(datos)

# Mostramos las primeras 5 filas
print("Primeras 5 filas:")
print(df.head())

# Mostramos la información general del DataFrame
print("\nInformación general:")
df.info()

# Calculamos el promedio general de todos los estudiantes
promedio_general = df["promedio"].mean()
print("\nPromedio general de calificaciones:", promedio_general)

# Mostramos solo los estudiantes con promedio mayor o igual a 80
print("\nEstudiantes con promedio mayor o igual a 80:")
print(df[df["promedio"] >= 80])

# Ordenamos de mayor a menor promedio
print("\nEstudiantes ordenados por promedio (mayor a menor):")
print(df.sort_values("promedio", ascending=False))
```

```python
**Explicación de métodos pandas:**
- `pd.DataFrame()` convierte un diccionario en una tabla organizada en filas y columnas.
- `head()` muestra solo las primeras 5 filas, para revisar rápido cómo quedaron los datos.
- `info()` muestra un resumen del DataFrame: cuántas filas hay, qué tipo de dato tiene cada columna y si hay valores vacíos.
- `mean()` calcula el promedio de todos los valores de una columna.
- `df[df["promedio"] >= 80]` filtra la tabla y muestra solo las filas que cumplen esa condición.
- `sort_values()` ordena la tabla según una columna. Con `ascending=False` ordena de mayor a menor.
```

```python
## Ejercicio 17: Análisis de ventas con pandas 
**Enunciado:**
- Cree un DataFrame de ventas con al menos 15 registros. 
- Incluya fecha, vendedor, producto, categoría, cantidad y precio unitario. 
- Calcule una nueva columna TotalVenta. 
- Calcule el total vendido por vendedor. 
- Identifique el producto más vendido. 
- Identifique la categoría con mayor ingreso. 
- Ordene las ventas de mayor a menor.

**Explicación de la lógica:**
- pd.DataFrame(datos)  crea una tabla de datos usando pandas.
- ventas["TotalVenta"] crea una nueva columna multiplicando cantidad por precio unitario.
- groupby() agrupa datos por una columna.
- sum() suma valores numéricos.
- idxmax() devuelve el elemento con el valor máximo.
- sort_values() ordena los datos.
ascending=False ordena de mayor a menor.
```

```python
import pandas as pd

# Datos de ventas
datos = {
    "Fecha": [
        "2025-05-01","2025-05-02","2025-05-03","2025-05-04","2025-05-05",
        "2025-05-06","2025-05-07","2025-05-08","2025-05-09","2025-05-10",
        "2025-05-11","2025-05-12","2025-05-13","2025-05-14","2025-05-15"
    ],

    "Vendedor": [
        "Ana","Luis","Pedro","Ana","Luis",
        "Pedro","Ana","Luis","Pedro","Ana",
        "Luis","Pedro","Ana","Luis","Pedro"
    ],

    "Producto": [
        "Laptop","Mouse","Teclado","Monitor","Laptop",
        "Mouse","Teclado","Laptop","Monitor","Mouse",
        "Laptop","Teclado","Monitor","Mouse","Laptop"
    ],

    "Categoria": [
        "Tecnologia","Accesorios","Accesorios","Tecnologia","Tecnologia",
        "Accesorios","Accesorios","Tecnologia","Tecnologia","Accesorios",
        "Tecnologia","Accesorios","Tecnologia","Accesorios","Tecnologia"
    ],

    "Cantidad": [
        2,5,3,1,2,
        4,2,1,2,3,
        2,4,1,6,3
    ],

    "PrecioUnitario": [
        800,20,35,250,800,
        20,35,800,250,20,
        800,35,250,20,800
    ]
}

# Crear DataFrame
ventas = pd.DataFrame(datos)

# Nueva columna TotalVenta
ventas["TotalVenta"] = ventas["Cantidad"] * ventas["PrecioUnitario"]

print("DataFrame completo")
print(ventas)

# Total vendido por vendedor
total_vendedor = ventas.groupby("Vendedor")["TotalVenta"].sum()

print("\nTotal vendido por vendedor")
print(total_vendedor)

# Producto más vendido
producto_mas_vendido = ventas.groupby("Producto")["Cantidad"].sum()

print("\nProducto más vendido")
print(producto_mas_vendido.idxmax())

# Categoría con mayor ingreso
categoria_ingreso = ventas.groupby("Categoria")["TotalVenta"].sum()

print("\nCategoría con mayor ingreso")
print(categoria_ingreso.idxmax())

# Ordenar ventas
ventas_ordenadas = ventas.sort_values(
    by="TotalVenta",
    ascending=False
)

print("\nVentas ordenadas")
print(ventas_ordenadas)
```

```python
---
## Ejercicio 18: Limpieza de datos con pandas
Enunciado: Crear un DataFrame con datos incompletos de clientes. Identificar valores nulos, contar nulos por columna, reemplazar ciudades vacías y eliminar registros sin correo.

Lógica: Se usan listas y ciclos para revisar datos vacíos. Luego se reemplazan ciudades vacías y se eliminan clientes que no tienen correo.
```

```python
clientes = [
    ["Ana", 22, "ana@gmail.com", "Tegucigalpa", "9999-1111"],
    ["Carlos", 35, "", "", "8888-2222"],
    ["María", "", "maria@gmail.com", "San Pedro Sula", ""],
    ["José", 28, "jose@gmail.com", "", "7777-3333"],
    ["Lucía", 40, "", "La Ceiba", "6666-4444"]
]

print("DATOS ORIGINALES")
for cliente in clientes:
    print(cliente)

# Contar valores vacíos
nulos = 0

for cliente in clientes:
    for dato in cliente:
        if dato == "":
            nulos += 1

print("\nCantidad de valores vacíos:", nulos)

# Reemplazar ciudades vacías
for cliente in clientes:
    if cliente[3] == "":
        cliente[3] = "No especificada"

# Eliminar clientes sin correo
clientes_limpios = []

for cliente in clientes:
    if cliente[2] != "":
        clientes_limpios.append(cliente)

print("\nDATOS LIMPIOS")

for cliente in clientes_limpios:
    print(cliente)
```

```python
Ejercicio 19: Análisis de préstamos

Enunciado: Crear un DataFrame con solicitudes de préstamo que incluya cliente, monto solicitado,
plazo, ingreso mensual y estado de solicitud. Mostrar las primeras 5 filas, la información general, 
el monto promedio solicitado, contar solicitudes por estado, mostrar únicamente préstamos aprobados, 
clientes con monto mayor a 50,000 y agrupar por estado calculando el promedio del monto solicitado.
```

```python
# Importamos pandas
import pandas as pd

# Datos de los prestamos
datos = {
    "cliente": ["Ana Lopez", "Carlos Perez", "Maria Gomez", "Jose Ruiz",
                "Lucia Torres", "Pedro Martinez", "Sofia Castro", "Luis Herrera"],

    "monto_solicitado": [45000, 60000, 30000, 80000,
                         52000, 47000, 90000, 25000],

    "plazo": [12, 24, 18, 36, 24, 12, 48, 6],

    "ingreso_mensual": [15000, 25000, 12000, 40000,
                        22000, 18000, 50000, 10000],

    "estado": ["Aprobado", "Pendiente", "Rechazado", "Aprobado",
               "Aprobado", "Pendiente", "Aprobado", "Rechazado"]
} 

# Crear el DataFrame
df = pd.DataFrame(datos)

# Mostrar las primeras filas
print("Primeras 5 filas:")
print(df.head())

# Informacion general
print("\nInformacion general:")
print(df.info())

# Promedio del monto solicitado
promedio = df["monto_solicitado"].mean()

print("\nPromedio del monto solicitado:", promedio)

# Cantidad de solicitudes por estado
print("\nSolicitudes por estado:")
print(df["estado"].value_counts())

# Solo prestamos aprobados
print("\nPrestamos aprobados:")
print(df[df["estado"] == "Aprobado"])

# Clientes con monto mayor a 50000
print("\nClientes con monto mayor a 50000:")
print(df[df["monto_solicitado"] > 50000])

# Agrupar por estado y sacar promedio
print("\nPromedio por estado:")
print(df.groupby("estado")["monto_solicitado"].mean())
```

```python
# Ejercicio 20: Mini análisis de datos tipo empresa
```

```python
# importar pandas
import pandas as pd

# crear los datos de empleados
datos = {
    "Nombre": ["Juan", "Maria", "Carlos", "Ana", "Luis", "Sofia"],
    "Departamento": ["TI", "Ventas", "Recursos Humanos", "TI", "Ventas", "Finanzas"],
    "Cargo": ["Programador", "Vendedor", "Analista", "Administrador", "Supervisor", "Contadora"],
    "Salario": [25000, 18000, 22000, 30000, 27000, 24000],
    "Experiencia": [2, 6, 4, 8, 10, 5],
    "Ciudad": ["Tegucigalpa", "San Pedro Sula", "La Ceiba", "Tegucigalpa", "Choloma", "Comayagua"]
}

# convertir a DataFrame
df = pd.DataFrame(datos)

# mostrar los primeros registros
print("Primeros registros")
print(df.head())

# estadisticas de salarios
print("\nEstadisticas de salarios")
print(df["Salario"].describe())

# promedio de salario por departamento
print("\nPromedio de salario por departamento")
print(df.groupby("Departamento")["Salario"].mean())

# empleado con salario mas alto
print("\nEmpleado con salario mas alto")
print(df[df["Salario"] == df["Salario"].max()])

# empleados con mas de 5 años de experiencia
print("\nEmpleados con mas de 5 años de experiencia")
print(df[df["Experiencia"] > 5])

# ordenar empleados por salario
print("\nEmpleados ordenados por salario")
print(df.sort_values(by="Salario", ascending=False))

# crear columna nivel de experiencia
niveles = []

for años in df["Experiencia"]:
    
    if años <= 3:
        niveles.append("Junior")
        
    elif años <= 7:
        niveles.append("Intermedio")
        
    else:
        niveles.append("Senior")

df["NivelExperiencia"] = niveles

# mostrar dataframe final
print("\nDataFrame final")
print(df)
```
